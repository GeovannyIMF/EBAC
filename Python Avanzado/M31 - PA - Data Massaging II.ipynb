## 6. Generar una lambda que añada la zona horaria y la capital del país de nacionalidad

```python
# Crear nuevo dataframe para obtener los datos de los paises únicos
countries = pd.DataFrame(df["Nationality"].unique(), columns=["Nationality"])
countries
```

```python
# Importar el módulo CountryInfo para obtener información de los paises
from countryinfo import CountryInfo

# Función para obtener la capital del país especificado
def getCapital(countryName):
    try:
        capital = CountryInfo(countryName).capital()
    except:
        capital = ""
    return capital

# Función para obtener las zonas horarias del país especificado
def getTimeZones(countryName):
    try:
        timezones = CountryInfo(countryName).timezones() 
    except:
        timezones = ""
    return timezones

# Aplicar la función lambda a la columna nationality para obtener las zonas horarias del país especificado
countries["TimeZones"] = countries["Nationality"].apply(lambda x: getTimeZones(x))

# Aplicar la función lambda a la columna nationality para obtener la capital del país especificado
countries["Capital"] = countries["Nationality"].apply(lambda x: getCapital(x))

#Imprimir el dataframe de paises con los datos obtenidos
countries
```

```python
# Combinar el dataframe de jugadores y países para agregar las columnas de zonas horarias y capital
df = df.merge(countries, how="inner", on="Nationality")
df
```
