## 5. Añadir una columna “Years Playing”, que calcule el año actual menos la columna “Joined”

```python
# Se utiliza la función now de datime para obtener la fecha y year para obtener el año
from datetime import datetime
current_year = datetime.now().year
current_year
```

```python
# Agregar columna con valor calculado del año actual menos la columna Joined
df["Years Playing"] = current_year - df["Joined"]
```

```python
# Verificar que la nueva columna fue agregada al dataframe
df.sample(5)
```

## 8. Generar un nuevo dataset que contenga el año (joined) y el número de jugadores (groupby)

```python
# Se utiliza groupby para agrupar los datos por la columna joined y se crea el nuevo dataset
new_df = df.groupby("Joined")["ID"].count().reset_index().rename(columns={"Joined":"Año", "ID":"Número de jugadores"})
new_df
```
