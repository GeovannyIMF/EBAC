#### ***** Ventas: *****

Todos estos números desglosados por ingreso: NA, EU, JP, Otros y el Global

**1) Total de ventas por año**

```python
# Agrupar datos por año y realizar la suma de cada columna de ventas
df.groupby(["Year"])[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].sum()
```

**2) Juegos más vendidos por año**

```python
# Agrupar datos por año y nombre y obtener los totales de ventas
df_year_name = df.groupby(["Year", "Name"])[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].sum()

#Ordenar por año en forma ascendente y despues por ventas en forma descendente
df_year_name = df_year_name.sort_values(by=["Year", "Global_Sales"],ascending=[True, False])

# Obtener el top 1 por cada año
df_year_name.groupby("Year").head(1).reset_index()
```

**3) Plataformas con más ingresos por año**

```python
# Agrupar datos por año y plataforma, obtener totales de ventas
df_year_platform = df.groupby(["Year", "Platform"])[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales", "Global_Sales"]].sum()

#Ordenar por año en forma ascendente y despues por ventas en forma descendente
df_year_platform = df_year_platform.sort_values(by=["Year", "Global_Sales"],ascending=[True, False])

# Obtener el top 1 por cada año
df_year_platform.groupby("Year").head(1).reset_index()
```

**4) Gráfico de comparación de ventas entre géneros por año**

```python
# crear tabla con genero como columnas, años como filas y ventas totales como valor
df_year_genre = pd.crosstab(df["Year"], df["Genre"], values=df["Global_Sales"], aggfunc="sum")

# Llenar valor nulos con 0
df_year_genre = df_year_genre.replace(np.nan, 0)
df_year_genre
```

```python
# Crear gráfico de barras apiladas
df_year_genre.plot(kind="bar", stacked=True, figsize=(10, 6))

# Agregar etiquetas y título
plt.title("Comparación de ventas entre géneros por año")
plt.xlabel("Año")
plt.ylabel("Total de Ventas")
plt.legend(title="Género")

# Mostrar gráfico
plt.show()
```

#### ***** Insights: 2 Gráficos adicionales interesantes: *****

**1) Top 5 de géneros más vendidos**

```python
# Agrupar por género, obtener las ventas totales por cada género
df_genre_total = df.groupby("Genre")["Global_Sales"].sum()

# Top 5 de Género mas vendidos
top_5_genre = df_genre_total.sort_values(ascending=False).head(5).iloc[::-1]

# Crear gráfico de barras horizontales
top_5_genre.plot.barh()

# Agregar etiquetas y título
plt.title("Top 5 de géneros mas vendidos")
plt.xlabel("Total de Ventas")
plt.ylabel("Géneros")

# Mostrar gráfico
plt.show()
```

**2) Gráfico de tendencia de ventas**

```python
# Agrupar por año, obtener las ventas totales por cada año
df_year_total = df.groupby("Year")["Global_Sales"].sum()

# Crear gráfico de línea o tendencia
df_year_total.plot()

# Agregar etiquetas y título
plt.title("Tendencia de ventas")
plt.xlabel("Año")
plt.ylabel("Total de Ventas")

# Mostrar gráfico
plt.show()
```

```python
# Crear gráfico de barras para el total de ventas por año
df_year_total.plot(kind="bar", stacked=False, figsize=(10, 6))

# Agregar etiquetas y título
plt.title("Total de ventas por año")
plt.xlabel("Año")
plt.ylabel("Total de Ventas")

# Mostrar gráfico
plt.show()
```

### 3. Revisar el EDA del ejercicio anterior y añadir siguientes elementos, basados en los últimos 4 años

#### ***** Crecimiento: *****

**1) qué género ha tenido el mayor y menor crecimiento**

```python
# Crear tabla con genero como columnas, años como filas y ventas totales como valor
df_year_genre = pd.pivot_table(df, index="Year", columns="Genre", values="Global_Sales", aggfunc="sum")

# Obtener los datos de los últimos cuatro años
df_last_years_genre = df_year_genre[-4:]

# Llenar datos nulos con valor 0
df_last_years_genre = df_last_years_genre.fillna(0)

# Obtener el porcentaje de cambio o crecimiento por género
crecimiento_por_genero = df_last_years_genre.pct_change() * 100

# Reemplazar los valores inf y -100 por NaN, debido a que estos cálculos no son correctos
crecimiento_por_genero = crecimiento_por_genero.replace([np.inf, -np.inf, -100], np.nan)
```

```python
# Género con el máximo crecimiento
# se obtiene el maximo de cada columna (Máximo de cada Género) y después se obtiene el máximo de las filas (Género con máximo crecimiento)
crecimiento_por_genero.max().idxmax()
```

```python
# Género con el mínimo crecimiento
# se obtiene el mínimo de cada columna (Mínimo de cada Género) y después se obtiene el Mínimo de las filas (Género con mínimo crecimiento)
crecimiento_por_genero.min().idxmin()
```
